```python
from google.colab import drive
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

#Mounting Drive
drive.mount('/content/drive')

spark=SparkSession.builder.appName("Mounting from drive").master("local[*]").getOrCreate()
df=spark.read.option("header","true").csv('/content/drive/MyDrive/csv files/coin_Bitcoin.csv')
df.show()
```
